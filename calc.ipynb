# In this sheet we calculate the parameters of a Gaussian beam from beam parameters $\alpha$

<font size="4">
Basic expressions:

waist: $w_0=\sqrt{\frac{\lambda z_R}{\pi}}$

    
Rayleigh length: $z_R=\frac{\pi w_0^2}{\lambda}$
    
Beam radius: $R(z)=z\left[1+\left(\frac{z_R}{z}\right)^2\right]$
    
waist $z$ dependence: $w(z)=w_0\left[1+\left(\frac{z}{z_R}\right)^2\right]^{1/2}$
    
if far field, the half angle divergence: $\vartheta_{FF}=\frac{\lambda}{\pi w_0}$
    
The source of the above expressions is: S.A. Self: *Focusing a spherical Gaussian beam*, Applied Optics **22**, 658 (1983).

</font>

<font size="4">Two parameters determine the beam, for example $R(z_1)$, $R(z_2)$; $w(z_1)$, $w(z_2)$; $w(z)$, $R(z)$; $w(z)$, $\vartheta_{FF}$
    </font>

<font size="4">Below we assume that $R(z_1)$ and $R(z_2)$ are known, $d=z_1-z_2$ </font>

```python
R1 = 2900
R2 = 2700
d = 820
z2 = d*(R1-d)/(R2-R1+2*d)
z1 = z2+d
lam = 0.98e-3
```

```python
print(f"z1 = {z1}, z2 = {z2}")
```

```python
z_R = np.sqrt(z1*(R1-z1))
w0 = np.sqrt(lam*z_R/np.pi)
print(f"z_R = {z_R}, w0 = {w0}")
```

<font size="4">Beam focusing:</font>

<font size="4">
Let us introduce a complex parameter that characterizes the Gaussian beam
    \medskip
    \begin{equation}
        \frac{1}{q}=\frac{1}{R(z)}-i\frac{\lambda}{\pi w^2(z)}=\frac{1}{z+i z_R}
    \end{equation}
</font>

<font size="4">
If we focus the beam with a thin lens of focal length $f$, then parameter $q$ transforms as
    \medskip
    \begin{equation}
        \frac{1}{q'}=\frac{1}{q}-\frac{1}{f}
    \end{equation}
</font>

<font size="4">
Separating the real and imaginary parts we obtain two equations. The real part gives
\begin{equation}
    (-z')-f=\frac{f^2(z-f)}{(z-f)^2+z_R^2}
\end{equation}
or in other form
\begin{equation}
     \frac{1}{{z/f}+(z_R/f)^2/\left(z/f-1\right)}+\frac{1}{-z'/f}=1,
\end{equation}
Note that $z$ and $z'$ are measured in the same coordinate system. Often they are measured in different coordinate systems, the origins are at the lens and both increase toward the focal point.
From the imaginary part  we get
\begin{equation}
        \frac{w_0'}{w_0}=\frac{f}{\left[(z-f)^2+z_R^2\right]^{1/2}}
\end{equation}
</font>

```python
import matplotlib.pyplot as plt
%matplotlib ipympl
from mpl_interactions import ipyplot as iplt

fig, ax = plt.subplots()
ax.grid()
ax.set_ylim(0,0.01)

f = 200
# Define function in a way you can re-use in calculations
s = np.linspace(-3000, -500, num=2500)

def focus(s, f):
    # sp = -f-f*f*(s-f)/( (s-f)**2+(z_R)**2 )
    wp0 = w0*f/np.sqrt((s-f)**2+(z_R)**2)
    theta = lam/(np.pi*wp0)
    return theta
    
    
ctrls = iplt.plot(s, focus, f=(100,350) )
```

```python
z0=np.pi*w0**2/0.98e-3
print(z0)
```

```python
def Rad(z):
    return z*(1+(z0/z)**2)
```
